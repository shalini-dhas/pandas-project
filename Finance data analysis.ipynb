*Visualization

```python
plt.style.use('ggplot')
df3['gender'].value_counts().plot(kind='bar', figsize=(8,5), ylabel='Count of People', xlabel='Gender',edgecolor = 'y', width=0.8,hatch='/')
sns.color_palette("mako", as_cmap=True)
plt.xlabel('gender-wise')
plt.ylabel('people count')
plt.title('Number of male and female investers', size = '13')
plt.legend()
plt.show()
```

```python
sns.countplot(x=df3['age'], hue=df3['gender'], palette='crest', edgecolor='r',hatch='/')
plt.legend(loc='upper right')
plt.xlabel('AGE-WISE')
plt.ylabel('people count')
plt.title('Investers by age', size = '13')
plt.show()
```

```python
plt.figure(figsize=(12, 7)) 
sns.countplot(x=df3['gender'],hue=df3['What are your savings objectives?'],linewidth=3,edgecolor='black',hatch='/')
plt.xlabel('gender_wise')
plt.title("Purpose behind the investments by mens & womens")
plt.show()
```

```python
df3['Factor'].value_counts().plot(kind='bar')
plt.ylabel('counts')
plt.title('Factor of Investment',size=15)
plt.show()
```

```python
fig, axes = plt.subplots(1, 2, figsize=(15,5))

sns.countplot(ax=axes[0], x=df3['Duration'])
axes[0].set_title('keep your money in any investment')

sns.countplot(ax=axes[1], x=df3['Invest_Monitor'])
axes[1].set_title('Investment_Monitoring')
```

```python
sns.countplot(x=df3['Duration'], hue=df3['Invest_Monitor'], palette='dark', edgecolor='black',hatch='/')
plt.legend(loc='upper right')
plt.xlabel('time period')
plt.ylabel('count of people')
plt.title('keep your money in investment')
plt.show()
```

```python
sns.countplot(x=df3['Source'])
plt.xlabel('Source')
plt.ylabel('investment')
plt.title('Source of information for investment')
plt.show()
```
